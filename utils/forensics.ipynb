```python
from lxml import etree

# Parse the XDMF file
tree = etree.parse("../dataset/h5_files/diffusion.xdmf")
root = tree.getroot()

# Print the root element and its children to inspect the structure
def print_elements(elem, level=0):
    print("  " * level + f"{elem.tag}: {elem.attrib}")
    for child in elem:
        print_elements(child, level + 1)

print_elements(root)
```
